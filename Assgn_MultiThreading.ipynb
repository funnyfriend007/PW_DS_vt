```python
### activeCount() is deprecated now, its replaced with active_count
### currentThread() is deprecated now, its replaced with current_thread

'''The threading.active_count() is an inbuilt method of the threading module, 
it is used to return the number of Thread objects that are active at any instant.
It comes with threading module
I Got this warning with activeCount
/tmp/ipykernel_77/165773695.py:14: DeprecationWarning: activeCount() is deprecated, use active_count() instead
'''

'''
The threading.enumerate() is an inbuilt method of the threading module, it is used to return the list of all the 
Thread class objects which are currently alive. It also includes daemonic threads, the main thread, 
and dummy thread objects created by current_thread(). It does not count the threads that have terminated or which have not started yet.
'''

'''
The threading.current_thread() is an inbuilt method of the threading module, 
it is used to return the current Thread object, which corresponds to the caller's thread of control.
'''

## threading.active_count , threading.eumerate method
try:
    import threading
    import time
    from functools import reduce
    def addlist(ls):
        print(reduce(lambda x,y: x+y ,ls))
    print("Number of active threads in the starting:", threading.active_count())
    print("Threads ============ ",*threading.enumerate(), sep = "\n")
    print(" ****** Active current thread right now 111111:", (threading.current_thread()))
     ## this will create thread objects         
    thread_lst_prime=[threading.Thread(target=addlist,args=(i,)) for i in [[1,3,5,7,9],[2,4,6,8,10],[1,3,5,7]]] 
    print(thread_lst_prime)     ## this will print thread names 
    print("active thread count after creatig new threads is ",threading.active_count())
    print("Threads NOW =========== ",*threading.enumerate(), sep = "\n")
    print(" ****** Active current thread right now 2222222:", (threading.current_thread()))
    for t in thread_lst_prime:  ## for every thread
        t.start()               ## starting thread one by one
        time.sleep(5)           ## sleep of 5 seconds 
        print("active thread count is ",threading.active_count())  ## checking active thread count
        print("Threads AGAIN =========== ",*threading.enumerate(), sep = "\n")
        print(" ****** Active current thread right now 3333333 :", (threading.current_thread()))
except Exception as e:
    print("Error is",e)
finally:
    print("program finished")
```

### Q4 - write a python to create 2 threads. Thread1 - print list of squares Thread2- print list of cubes

```python
try:
    import threading
    def func(op):
        if op=="SQ":
            print([x**2 for x in range(25)])
        elif op=="CU":
            print([x**3 for x in range(25)])
    thrd=[threading.Thread(target=func,args=(i,)) for i in ["SQ","CU"]]
    for t in thrd:
        print(t)
        t.start()
except Exception as e:
    print("Exception is ",e)
```

### Q6 - explain deadlocks & race conditions 

```python
### Race condition 
'''A race condition is defined as a scenario when two or more threads access the shared 
resources with write permission and try to change the data. Thus, the value of such variables becomes unpredictable.
When a thread is about to execute the critical section, it can acquire the lock by calling the acquire() function. 
It can then execute the critical section and release the lock by calling the release() function on the lock.
In case of global/shared variables we can use join method else acquire & release
'''
##for example - 
try:
    #  race condition with a shared variable and join method
    import threading
    # make additions into the global variable
    def addnum(amount,i):
        global value
        for _ in range(i):
            value+=amount
    # define the global variable
    value = 0
    # start a thread making additions
    add_thr = threading.Thread(target=addnum, args=(100,1000000))
    add_thr.start()
    # wait for both threads to finish
    print('Waiting for threads to finish... ')
    add_thr.join()
    # print the value only when thread is finished
    print(f'Value is: {value}')
except Exception as e1:
    print(e1)

## Deadlock 
'''Deadlock: The program cannot continue to run because the lock of a thread is not released in time. It refers to a phenomenon in 
which two or more processes or threads are waiting for each other due to competition for resources during the execution process. 
If there is no external force, they will not be able to advance. At this time, the system is said to be in a 
deadlock state or the system has a deadlock, and these processes that are always waiting for each other are called deadlock processes.
'''
```
